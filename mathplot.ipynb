```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,6])
y=np.array([4,10])
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y1=[ np.sin(i) for i in x]
y2=[ np.cos(i) for i in x]
plt.plot(x, y1,)
plt.plot(x, y2)
plt.xlabel("Angles")
plt.ylabel("sin/cos(x)")
plt.legend(["sin", "cos"], loc ="upper left")
plt.ylim(-1.5, 2.0)
```
