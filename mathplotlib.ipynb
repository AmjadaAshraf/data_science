```python
#sin and cs value upto limit 7np
import matplotlib.pyplot as plt
import numpy as p
x=p.arange(0,5*p.pi,0.1)
s=p.sin(x)
c=p.cos(x)
plt.plot(x,s,x,c)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as p
dta={'anju':50, 'anila':60, 'anu':70}
studentname=list(dta.keys())
totalmarks=list(dta.values())
plt.bar(studentname,totalmarks,color='blue',width=0.5)
plt.xlabel("STUDENTNAME")
plt.ylabel("TOTALMARKS")
plt.title("STUDENT DETAILS")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as p
x=p.array([1,2,3,4])
y=[i ** 2 for i in x]
plt.xlabel("numbers")
plt.ylabel("squares")
plt.scatter(x,y)
plt.xlim(1,5)
plt.ylim(1,4)
plt.plot(x,y)
plt.legend()
plt.show()
```
